```python
import re, string

def clean_str(string):
  """
  String cleaning before vectorization
  """
  try:    
    string = re.sub(r'^https?:\/\/<>.*[\r\n]*', '', string, flags=re.MULTILINE) #removing special characters
    string = re.sub(r"[^A-Za-z]", " ", string)         
    words = string.strip().lower().split()    #removing unwanted spaces and convertingto lower case
    words = [w for w in words if len(w)>=1]
    return " ".join(words)	
  except:
    return ""
```

```python
df.sypnopsis
```

```python
df['synopsis'] = df['sypnopsis'].apply(clean_str)
df.synopsis
```

```python
df['Name'] = df['Name'].apply(clean_str)
```

```python
df['tags']=df.synopsis+df.Genres
df.head()
```

```python
ndf=df[['MAL_ID','Name','Score','tags']]
ndf.head()
```

```python
ndf=ndf[ndf.Score != 'Unknown']
ndf.shape
```

```python
ndf=ndf.sort_values(by='Score',ascending=False)
ndf=ndf.iloc[:5001,:]
ndf.shape
```

```python
# similarity on the basis of tags, members or ratings
# similarity between tags ?
# we will convert text to vectors using bag of words as it is simple
# closest vectors we recommend

# add all the tags and check frequency of each word and extract frequency
# of top 5000 words
# so you get a table with number of anime vs 5000 top words 
# we won't be using stop words
```

```python
import nltk 
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
ndf.tags=ndf.tags.apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(ndf['tags']).toarray()
vectors.shape
```

```python
cv.get_feature_names()
```

```python
# calculate distance between movie distance using cosine distance that is 
# anglebetween them as they are vectors
# distance is inversely related similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
similarity
```

```python
similarity.shape
```

```python
#fetch_index
#sort distances 
# pick 5 closest anime

# enumerate to lock index values
```

```python
def recommend(anime):
    anime_index=ndf[ndf['Name']==anime].index[0]
    distances=similarity[anime_index]
    anime_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in anime_list:
        print(ndf.iloc[i[0]].Name)
```

```python
ndf.head()
```

```python
recommend('cowboy bebop')
```
